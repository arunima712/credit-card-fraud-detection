```python
df.isnull().sum()
```

```python
df= df.drop(['CUST_ID'], axis = 1)

df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X=scaler.fit_transform(df)
```

```python
print(X.shape)
```

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(0.95)
```

```python
X_pca= pca.fit_transform(X)
```

```python
X.shape
```

```python
X_pca.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model= KMeans(n_clusters=18)
```

```python
model.fit_predict(X_pca)
```

```python
model.inertia_
```

```python
inr_list = []
```

```python
for i in range(1,20):
  model = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )
  model.fit(X_pca)
  inr= model.inertia_
  inr_list.append(inr)
```

```python
print(inr_list)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(range(1, 20), inr_list, marker = 'o')

plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for K-Means: Number of Clusters vs. Inertia')
plt.grid()

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(X_pca)

df['Cluster'] = clusters
```

```python
pca_vis = PCA(n_components=2)
X_vis = pca_vis.fit_transform(X)

plt.figure(figsize=(10,6))
sns.scatterplot(
    x=X_vis[:,0],
    y=X_vis[:,1],
    hue=clusters,
    palette='Set1'
)

plt.title("Customer Segmentation using PCA + KMeans")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_pca, clusters)
print("Silhouette Score:", score)
```

```python
df['Cluster'] = clusters

cluster_summary = df.groupby('Cluster').mean()
print(cluster_summary)
```
